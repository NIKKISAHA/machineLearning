```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,5,11)
y=x**2
plt.plot(x,y)
```

```python
plt.plot(x,y,"r")
plt.xlabel("x")
plt.ylabel("y")
plt.title("pyplot")
```

```python
plt.subplot(1,2,1).plot(x,y,"r")
plt.subplot(1,2,2).plot(x,y,"b")
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.2,0.3,0.2])
axes.plot(x,y,"b")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("pyplot")
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("pyplot")
fig.savefig("secondpic.png")
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.2,0.3,0.2])
axes.plot(x,y,"b",label="xy")
axes.legend()
```

```python
plt.plot(x,y,"g--",label="xy")
plt.hist([1,2,3])
plt.scatter(x,y)
plt.legend()
```
