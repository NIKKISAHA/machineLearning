```python
from scipy.optimize import linear_sum_assignment
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle

data =pd.read_csv("/content/drive/MyDrive/dataset/california_housing_test.csv")
# data.head(5)
data = data[["longitude","latitude","housing_median_age","households","population"]]
pridict= "population"
# data
x=np.array(data.drop([pridict],1))
y=np.array(data[pridict])
# x
# y
x_train,x_test,y_train,y_test = sklearn.model_selection.train_test_split(x,y,test_size=0.1)
linear=linear_model.LinearRegression()
linear.fit(x_train,y_train)
acc= linear.score(x_test,y_test)
# acc
# linear.coef_
# linear.intercept_
p=linear.predict(x_test)
for i in range(len(p)):
  print(p[i],x_test[i],y_test[i])
```
