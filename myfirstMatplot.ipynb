```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.figure(figsize=(8,5),dpi=100)
plt.title("me",fontdict={'fontsize':40})
plt.xlabel("x",fontdict={'fontsize':40})
plt.ylabel("y",fontdict={'fontsize':40})
plt.plot(x,y,'r+-',label="2x")
plt.savefig("myfirst2.png",dpi=300)
plt.show()
```

```python
from ast import pattern
l=['a','n','k']
v=[1,2,3]
plt.figure(figsize=(8,5),dpi=100)
bars=plt.bar(l,v)
pattern=['/','O','*']
for b in bars:
  b.set_hatch(pattern.pop(0))
plt.savefig("myfirst.png",dpi=300)
plt.show()
```
