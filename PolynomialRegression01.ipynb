```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

mydata=[(1,5000),(2,10000),(5,15000),(9,20000),(11,22000),(3,23000)]
mydf=pd.DataFrame(mydata,columns=["lavel","salary"])
# mydf

x=mydf.iloc[:,0:1].values
# x
y=mydf.iloc[:,1].values
# y

# plt.scatter(x,y)
sns.lmplot(x="lavel",y="salary",data = mydf)
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
reg.fit(x,y)
reg.predict([[6]])
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(2)
x_poly=poly.fit_transform(x)

reg2=linear_model.LinearRegression()
reg2.fit(x_poly,y)
reg2.predict(poly.fit_transform([[6]]))
```
